# 1. Write a Python Program to Check if a Number is Positive, Negative or Zero?

```python
a = int(input("Enter the value :"))

def number(a):
    if a == 0:
        print(a,"The Number is Zero")
    elif a > 0:
        print(a,"The Number is Positive")
    else :
        print(a,"The number is Negative")
    return a 
print(number(a))
```

# 2. Write a Python Program to Check if a Number is Odd or Even?

```python
number = int(input("Enter the value :"))

def odd_even(number):
    if (number%2) == 0:
        print(number,"The number is Even")
    else :
        print(number,"The number is Odd")
    return number
odd_even(number)
```

# 3. Write a Python Program to Check Leap Year?

```python
a = int(input("Enter any year that is to be checked for leap year : "))

def leap_year(a):
    if (a%4)== 0:
        print(a,"The given number is a Leap Year")
    else :
        print(a,"The given number is not a Leap Year")
    return a
leap_year(a)
```

# 4. Write a Python Program to Check Prime Number?

```python
num = int(input("Enter the desired number :"))

def is_prime(num):
    if num>1:
        for i in range(2,num):
            if(num%i) ==0 :
                print("The entered number is not a prime number ")
            else:
                print("The entered number is a prime number")
            break
    else:
        print("The entered number is not a prime number ")
    return num

is_prime(num)
```
